```python
import numpy as np

with open('./assets/iris.csv', encoding = 'utf') as f:
    data = np.loadtxt(f, dtype=float)
    #print(type(data))  #ndarray

    max_loc = np.argmax(data)
    min_loc = np.argmin(data)
    print('max_loc:\t',max_loc)
    print('min_loc:\t',min_loc)

    max = np.max(data)
    min = np.min(data)
    print('max:\t', max)
    print('min:\t', min)

    data_sorted = np.sort(data)
    print('data_sorted:\n',data_sorted)

    data_unique = np.unique(data_sorted)
    print('data_unique:\n', data_unique)

    data_mean = np.mean(data_sorted)
    print('mean:\t', data_mean)

    data_var = np.var(data_sorted)
    data_std =  np.std(data_sorted)
    print('data_var:\t',data_var)
    print('data_std:\t', data_std)

    data_sum = np.sum(data_sorted)
    print('data_sum:\t', data_sum)

    data_cumsum = np.cumsum(data_sorted)
    print('data_cumsum:\t', data_cumsum)

    count =( data_sorted > 5.84).sum()
    print('number of bigger than mean 5.84:\t', count)
```
