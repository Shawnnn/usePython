```python
import numpy as np 
import pandas as pd 

data = {
    'company': ['杭州银行', '青农银行', '常熟银行','工商银行','上海银行','江苏银行'],
    'code':['600926', '002958', '601128', '601398', '601229', '600919'],
    'marketcap':[449, 371, 237, 21313, 1369, 823],
    'PE':[8.31, 15.36, 16.01, 7.16, 7.59, 6.3]

}
df_company = pd.DataFrame(data)
print(df_company.head(5))
```

```python
dr = pd.date_range(start = '2019-01-02', periods = 100)
data = np.random.randn(100).cumsum()
close = data - np.min(data)
df = pd.DataFrame({'close':close}, index = dr)
df
```

```python
df['roll_mean_5'] = df.rolling(window=5).mean()
import matplotlib.pyplot as plt 
df[['close','roll_mean_5']].plot(subplots = True, figsize = (9,5), grid =True)
plt.show()
```
